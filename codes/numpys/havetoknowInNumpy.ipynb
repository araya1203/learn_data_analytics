```
import numpy as np
```

```
student_score = [
 [1, 90, 80, 85, 0, 0.0, 0, 0],
 [2, 88, 85, 90, 0, 0.0, 0, 0],
 [3, 32, 85, 90, 0, 0.0, 0, 0],
 [4, 42, 75, 93, 0, 0.0, 0, 0],
 [5, 82, 86, 88, 0, 0.0, 0, 0]
]
ndarray = np.array(student_score)
ndarray, type(ndarray), type(ndarray[0])
```

## Indexing & Slicing

```
ndarray[:, 0], ndarray[4, :], ndarray[:, 1:4]
```

## Statistic

```
np.max(ndarray[:, 0],axis=0), np.sum(ndarray[:,1:4],axis=1)
```

```
ndarray[:, 4] = np.sum(ndarray[:,1:4],axis=1)
ndarray, type(ndarray), ndarray.shape
```

## 조건연산(True, False)

```
(ndarray == 0), (ndarray[:,1:5] > 20), (ndarray[:,1:5] > 20) & (ndarray[:,1:5] > 80)
```

## Mask

```
ndarray[(ndarray == 0)]
```

```
# same Mask ndarray size
ndarr = ndarray[:,1:5]
ndarr[(ndarr > 20)]
```
