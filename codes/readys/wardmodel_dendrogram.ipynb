```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.cluster.hierarchy import ward, dendrogram
import matplotlib.pyplot as plt
import nltk

# nltk 데이터 다운로드 (한 번만 실행)
nltk.download('punkt')

# 댓글 데이터 생성 예시 (문자열 리스트 형태)
comments = [
    "이 영화 정말 재밌네요!",
    "진짜 최고의 영화입니다.",
    "별로였어요.",
    "내용이 너무 어렵습니다.",
    "배우들의 연기가 좋았습니다."
]

# 텍스트 전처리 및 벡터화
vectorizer = TfidfVectorizer(tokenizer=nltk.word_tokenize)
comment_vectors = vectorizer.fit_transform(comments)

# 와드 클러스터링 수행
linkage_matrix = ward(comment_vectors.toarray())

# 덴드로그램 시각화
dendrogram(linkage_matrix, labels=comments)

plt.xlabel('댓글 인덱스')
plt.ylabel('거리')
plt.xticks(rotation=90)  # x축 레이블 회전 (옵션)
plt.show()
```
