```python
import glob

filename_list = []
for file_name in glob.glob("./*.csv"):
    filename_list.append(file_name)
filename_list
```

##read csv from files and concat DataFrammes

```python
import pandas as pd
```

```python
df_csv_list = list()
for file_name in filename_list:
  df_csv_list.append(pd.read_csv(file_name))

df_concat = pd.concat(df_csv_list)
# df_concat
```

```python
# show all rows
pd.options.display.max_columns = df_concat.shape[1]
df_concat.head()
```

##경기도 시군구별 상가 현황

```python
df_concat.columns
```

### 시군구별 평균 위도 구하기

```python
stores_one_area = df_concat[df_concat['시도명'] == '경기도'].copy()
stores_one_area.head()
```

### 사전 시각화 : 커피전문점/카페/다방 

```python
stores_one_area_class_cafes = stores_one_area[stores_one_area['상권업종소분류명'] == '커피전문점/카페/다방']
```

####그래프 지도 그리기

```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.scatterplot(data=stores_one_area_class_cafes, x='경도', y='위도', hue="시군구명")
plt.legend(ncol=2, loc=(1, 0))

plt.show()
```

### 지도에 시군구명 표시

```python
stores_one_area_locatioin = stores_one_area[['시군구명', '경도', '위도']].groupby('시군구명').mean()
# stores_one_area_locatioin
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(data=stores_one_area_class_cafes, x='경도', y='위도', hue="시군구명", marker='.',
                alpha=0.3)
plt.legend(ncol=2, loc=(1, 0))

for i in range(len(stores_one_area_locatioin)):
  x = stores_one_area_locatioin.iloc[i,0]
  y = stores_one_area_locatioin.iloc[i,1]
  text = stores_one_area_locatioin.index[i]
  plt.text(x,y,text, ha='center', fontsize='large')

plt.title('Cafes')
plt.show()
```

###hexbin으로 그리기

```python
plt.figure(figsize=(20,15))
plt.hexbin(stores_one_area_class_cafes['경도'], stores_one_area_class_cafes['위도'], cmap='Blues', gridsize=30)
plt.colorbar()

for i in range(len(stores_one_area_locatioin)):
  x = stores_one_area_locatioin.iloc[i,0]
  y = stores_one_area_locatioin.iloc[i,1]
  text = stores_one_area_locatioin.index[i]
  plt.text(x,y,text, ha='center', fontsize='large')

plt.show()
```
