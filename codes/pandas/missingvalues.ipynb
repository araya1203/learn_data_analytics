```
import pandas as pd
import numpy as np
```

```
student_score_dict = [
{'name':'김철수','kor':90, 'eng':'', 'math':0},
{'name':'박제동','kor':70, 'eng':50, 'math':100},
{'name':'홍길동','kor':80, 'eng':None, 'math':''},
{'name':'김철수','kor':90, 'eng':20, 'math':70},
{'name':'장길산','kor':None, 'eng':120, 'math':None},
{'name':'홍길동','kor':80, 'eng':None, 'math':0},
{'name':'정진수','kor':90, 'eng':0, 'math':80},
{'name':'오금순','kor':50, 'eng':70, 'math':None},
{'name':'정만득','kor':None, 'eng':None, 'math':None},
]
df_data = pd.DataFrame(student_score_dict)
```

# Cechk duplicat

```
df_data.describe(include='all')
```

```
df_data['name'].duplicated()
```

```
df_data['name'].drop_duplicates(keep='first')   # keep='first', 'last', False argument
```

## drop by mask

```
mask_dup = df_data['name'].duplicated() & df_data['eng'].isnull()
mask_dup
```

```
df_data = df_data.drop(df_data[mask_dup].index, axis=0)
df_data
```

## Check Null

```
df_data.info(), df_data.isnull().sum(), df_data['name'].value_counts().head()
```

## Drop row with Null

```
df_data.dropna()      # df_data.dropna(axis=1)
```

## Fill column with something

```
df_data['kor'] = df_data['kor'].fillna(df_data['kor'].mean())
```

##Check string

```
df_data.dtypes, df_data.empty
```

```
# df_data['eng','math'] = df_data[['eng','math']].replace({'':np.nan})  # Error
df_data['eng'] = df_data['eng'].replace({'':np.nan})
df_data
```

```
df_data['eng'] = df_data['eng'].fillna(df_data['eng'].median())    # 중간값
```

```
df_data['math'].replace({'':np.nan}, inplace = True)
df_data['math'].fillna(df_data['math'].median(), inplace = True)    
# df_data['math'].fillna(df_data['math'].mode(), inplace = True)    # 최빈값
```

```
df_data
```

```
df_data = df_data.astype({"kor": int, "eng": int, "math": int})
df_data
```

```
df_data.describe().T
```

## get total

```
df_data['total'] = df_data['kor']+df_data['eng']+df_data['math']
df_data
```

```
df_data['total'] = df_data.iloc[:,1:4].sum(axis=1)
df_data
```
