```python
# credit card의 만료일이 2025년도인 카드를 소지한 고객수?
```

```python
# 도메인별 고객의 이메일 숫자?
```

```python
df['Email'][0].split('@')[1]
```

```python
f = lambda x: x.split('@')[1]
```

```python
df['Email'].apply(f).value_counts()
```

```python
df['email_domain'] = df['Email'].apply(f)
df.head()
```

```python
df.groupby('email_domain').Lot.count().sort_values(ascending=False)
```
