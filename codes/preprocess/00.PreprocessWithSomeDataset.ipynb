#Cross validation

```
import pandas
import numpy as np

dataset = pandas.read_csv('boston_df.csv')
X = dataset.drop(['RAD'], axis=1)
y = dataset['RAD']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)

from sklearn.model_selection import KFold
from sklearn.svm import SVC
best_svr = SVC(kernel='rbf')
scores = []
cv = KFold(n_splits=10, random_state=42, shuffle=True)
for train_index, test_index in cv.split(X):
    # print("Train Index: ", train_index, "\n")
    # print("Test Index: ", test_index)
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    best_svr.fit(X_train, y_train)
    scores.append(best_svr.score(X_test, y_test))

print('KFold mean score : ',np.mean(scores))
# best_svr = SVR(kernel='rbf')
from sklearn.model_selection import cross_validate
cv_score = cross_validate(best_svr, X, y, cv=10, scoring='f1_weighted')
# score = cross_val_predict(best_svr, X, y, cv=10)
print('cross_val_predict mean score : ',np.mean(cv_score['test_score']))   
# score
```

###Telco-Customer-Churn

```
import pandas as pd
telco_df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
# telco_df.head()
telco_df.
```

```
telco_df_y = telco_df['MonthlyCharges']
telco_df_x = telco_df[['tenure','gender', 'PaymentMethod']]
telco_df_x = pd.get_dummies(telco_df_x, columns=['gender', 'PaymentMethod'], drop_first=True)
telco_df_x.head()
```

####OLS with statmodel

```
import statsmodels.api as sm
import numpy as np
sm_constant = sm.add_constant(telco_df_x)
estimate = sm.OLS(telco_df_y, sm_constant)
results = estimate.fit()
print(results.summary())
```

####LinearRegression with sklearn

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(telco_df_x, telco_df_y)
lr.score(telco_df_x, telco_df_y)
```
