## make pandas

```
import pandas as pd
df = pd.read_csv("data.csv")
df
```

```
np_array = df.values
np_array
```

```
np_array[:,1:3]
```

##Imputer

```
import sklearn
sklearn.__version__
```

```
from sklearn.impute import SimpleImputer
```

```
# imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)  # old version
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(np_array[:,1:3])
```

```
np_array[:,1:3] = imputer.transform(np_array[:,1:3])
```

```
np_array
pd.to_csv('')
```

# Feature Scaling

```
from sklearn.preprocessing import StandardScaler
```

```
standardScaler = StandardScaler()
np_array[:,1:3] = standardScaler.fit_transform(np_array[:,1:3])
np_array
```
