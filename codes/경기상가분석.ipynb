```python
import glob

filename_list = []
for file_name in glob.glob("./*.csv"):
    filename_list.append(file_name)
filename_list
```

##read csv from files and concat DataFrammes

```python
import pandas as pd
```

```python
df_csv_list = list()
for file_name in filename_list:
  df_csv_list.append(pd.read_csv(file_name))

df_concat = pd.concat(df_csv_list)
# df_concat
```

```python
# show all rows
pd.options.display.max_columns = df_concat.shape[1]
df_concat.head()
```

##경기도 상가 현황

```python
df_concat.columns
```

```python
stores_one_area = df_concat[df_concat['시도명'] == '경기도'].copy()
stores_one_area.head()
```

```python
stores_one_area_series = stores_one_area['시군구명'].value_counts()
stores_one_area_series
```

###그래프 그리기 : 판다스 사용

```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

plt.bar(stores_one_area_series.index, stores_one_area_series.values)
plt.show()
```

```python
# stores_one_area_series.plot?
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(121)
# stores_one_area_series.plot.bar()
stores_one_area_series.plot(kind='bar', color='r')
plt.subplot(122)
stores_one_area_series.plot.pie(autopct='%.1f%%', pctdistance=0.8)

fig.suptitle('경기도 상가 현황')
fig.tight_layout()
plt.show()
```

##지역별 / 업종 대분류 상가 현황

```python
stores_one_area_pivot = stores_one_area.pivot_table(index='상권업종대분류명', columns='시군구명', values='상호명', aggfunc='count')
stores_one_area_pivot
```

###heatmap 그리기

```python
import seaborn as sns
```

```python
plt.figure(figsize=(25,10))
sns.heatmap(stores_one_area_pivot, cmap='Blues', annot=True, fmt='')
plt.show()
```
