```python
# 데이터 필터(특정 조건을 만족하는 데이터를 추출 1oc : location )
cond1 =(df1['정지시간2'] >= 600)
df1.loc [cond1]
```

```python
# 두 조건을 모두 만족하는 데이터만 추출
cond1 =(df1['정지시간2'] >= 600)
cond2 =(df1['작업구분'] == '야간')
df1.loc[cond1 & cond2] # 두 조건을 모두 만족하는 데이터를 추출(8, and)
```

```python
# 두 조건 중 하나라도 만족하면 데이터를 추출
cond1 =(df1['정지시간2'] >= 600)
cond2 =(df1['메세지'] =='초기화 Error')
df1.loc[cond1 | cond2] # OR, /
```

```python
# COlE 8%(pivot_table)
df1.pivot_table( index= '메세지', values='정지시간2', aggfunc= 'mean')
# index ='구분하고 싶은 항목(범주형)'
# values ='계산하고자 하는 값(연속형)'
# aggfunc ='계산하고자 하는 동계량!
```

```python
# Mission 1-1.오류 메세지'메세지'항목에서'Door Open'에 해당하는 오류들 중에
# '정지시간2'이 가장 오래걸린 상위 100개 데이터를 추출
cond1 =(df1['메세지'] == 'Door Open')
df1.loc[cond1].sort_values(by='정지시간2',ascending=False).head(100)
```

```python
# Wission 1-2. ' 정자시간2'가 300초 이상이면서, '작업자'가 '이은미' 인 데이터를 추출해
# '정지시간인'의 요약통계량을 확인
cond1 = (df1['작업자'] =='이은미')
cond2 = (df1['정지시간2'] >= 300)
df1.loc[cond1 & cond2]
# 0001jmpegmai 1.com / 3명 / 던킨 / Code / 35 분
```
