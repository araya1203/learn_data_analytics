```python
 import pandas as pd # import: 라이브러리를 해당 스크립트로 호출하는 함수
```

```python
# COlE =497
df1 = pd.read_csv('preprocessing_data.csv')
# read csv : csv 파일을 불러오는 함수
df1.head(3) # head( ) : 상위 7개의 데이터를 확인
```

```python
df1.columns
```

```python
Y = df1['target_binom']
# drop() : 데이터 프레임에서 불필요한 항목을 제거
X = df1.drop(columns=[ 'Target'
, 'Error_message'
, 'target_binom', 'Chamber_Route'
, 'Unnamed: 0', 'Ox_Chamber'
, 'process', 'type'
, 'No_Die', 'Unnamed: 0_x', 'photo_soft_Chamber'
, 'Unnamed: 0_y'
, 'Etching_Chamber'
, 'Chamber_Num'
, 'process4'])
```

```python
# 숫자데이터만 선택
numeric_list = X.describe().columns.tolist()
X1 = df1[numeric_list ]
X1
```

### 간단히 적용 확인

```python
# 학습데이터와 검증데이터 분할
from sklearn.model_selection import train_test_split
# 학습데이터와 검증데이터를 8:2 비율로 분할 (Random Samp / ing)
X_train, X_test, Y_train, Y_test = train_test_split(X1,Y, random_state=1234)
# randon state : 무작위 추출을 원하는 Index로 맞추어 추출
print(X_train.shape) # shape : 데이터의 행과 열개수를 출력
print(X_test.shape)
```

```python
# 학습 수행
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier ()
model.fit(X_train, Y_train)
```

#### 결측치 (Missing Value)
# 1) 제거 : 결측값이 있는 행을 삭제
# 2) 대치 : 다른 값을 결측값에 대체 (평균 / 중앙값) -> 알고리증 (knn)

### 평가

```python
from sklearn.metrics import accuracy_score
# 정확도 : 전체 데이터 중 정확하게 분류한 데이터의 수 0~
Y_train_pred = model .predict (X_train)
accuracy_score(Y_train, Y_train_pred)
```

```python
Y_test_pred = model .predict (X_test)
accuracy_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision_score(Y_test, Y_test_pred)
```

```python
recall_score(Y_test, Y_test_pred)
```

```python
f1_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(Y_test, Y_test_pred)
print(report)
```
