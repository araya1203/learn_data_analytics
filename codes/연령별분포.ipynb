```python
import pandas as pd
read_df = pd.read_csv('./202106_202106_연령별인구현황_월간.csv', encoding='cp949')

read_df
```

##데이터 정리

###컬럼명 정리

```python
def replace_column_name(column_name, split_param='_'):
  split_list = column_name.split(split_param)
  result = split_list[len(split_list)-1]
  return result
```

```python
read_df.columns = list(map(replace_column_name,list(read_df.columns)))
```

```python
read_df
```

```python
read_df.columns
```

```python
#drop some columns
read_df.drop(columns=['총인구수', '연령구간인구수'], inplace=True)
```

###index 변경

```python
read_df.set_index('행정구역', inplace=True)
```

```python
read_df.dtypes
```

```python
(read_df.dtypes == 'int64').all()
```

```python
type(read_df['99세'].dtype), read_df['99세'].dtype == 'object'
```

```python
change_dtype_columns = read_df.columns
for column in change_dtype_columns:
  if read_df[column].dtype == 'object':
    read_df[column] = read_df[column].str.replace(',','').astype('int64')
read_df  
```

```python
(read_df.dtypes == 'int64').all()
```

##특정 동 인구 분석

###학군지 분석 : 대치동와 목동

```python
condition = read_df.index.str.contains('강남구 대치') | read_df.index.str.contains('양천구 목')
dong_df = read_df[condition].copy()
dong_df
```

```python
fig, ax = plt.subplots(3,3, figsize=(15,10), sharey=True)
rows, columns = 0, 0
for index in range(len(dong_df)):
  ax[rows][columns].plot(dong_df.iloc[0])
  ax[rows][columns].set_xticks(range(0, 101, 10))
  ax[rows][columns].set_title(dong_df.index[index].split(' ')[1])
  columns += 1
  if columns % 3 == 0:
    columns = 0
    rows += 1

fig.tight_layout()
```

###특정 동별 인구 분석

```python
specific_gu_df = read_df[read_df.index.str.contains('관악구')]
specific_gu_df
```

```python
fig, ax = plt.subplots(7,3, figsize=(15,10), sharey=True)
rows, columns = 0, 0
for index in range(1, len(specific_gu_df)):
  ax[rows][columns].plot(specific_gu_df.iloc[0])
  ax[rows][columns].set_xticks(range(0, 101, 10))
  ax[rows][columns].set_title(specific_gu_df.index[index].split(' ')[2].split('(')[0])
  columns += 1
  if columns % 3 == 0:
    columns = 0
    rows += 1

fig.tight_layout()
```
